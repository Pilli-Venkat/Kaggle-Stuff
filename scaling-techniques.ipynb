```python
import numpy as np
import pandas as pd
```

```python
X = np.array([1,2,3,1,2,1,2])
X= X.reshape(-1,1)
```

## 1) Standardization :
mean = 0 , std = 1


### Implementation 


```python
def Standard_Scaler(x):
    return (x - np.mean(x))/np.std(x)
print(Standard_Scaler(X))
```

### Scikit-learn

```python
from sklearn.preprocessing import StandardScaler

scaler_obj = StandardScaler()

print(scaler_obj.fit(X))
print(scaler_obj.transform(X))
```

## 2) Normalization :
Scales value 0 to 1

### Implemenation

```python
def Normalize(x):
    return ( x- min(x))/(max(x) - min(x))

print(Normalize(X))
```

### Scikit-Learn


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
norm_object = MinMaxScaler()
norm_object.fit(X)
print(norm_object.transform(X))
```
